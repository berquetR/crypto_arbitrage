```python
import requests
import os

base_url = "https://historical-data.kucoin.com/data/spot/daily/trades/SOLUSDT/"

for month in range(1, 12):  # From January (01) to November (11)
    month_str = str(month).zfill(2)

    for day in range(1, 32):  # Assuming up to 31 days in a month
        day_str = str(day).zfill(2)
        filename = f"SOLUSDT-trades-2023-{month_str}-{day_str}.zip"
        url = f"{base_url}{filename}"
        response = requests.get(url)

        if response.status_code == 200:
            save_path = f"../../data/KuCoin/{filename}"
            os.makedirs(os.path.dirname(save_path), exist_ok=True)

            with open(save_path, "wb") as f:
                f.write(response.content)
            print(f"Downloaded: {url}")
        else:
            print(f"Failed to download: {url}")
```
